```python
import pandas as pd
import numpy as np
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns


data = pd.read_csv('pitchers202302.csv')


#data = data[data.Gm > 10]
print (data)

#What game is it? How many innings has the pitcher pitched this year? what is their era going into the game, and how many
#days rest have they had?
x = data[['Gtm','TIP','TERA','DR']].values

#W or L
y = data['TRslt'].values

# Splitting dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)




        

```

```python
# Creating Decision Tree classifier object
clf = RandomForestClassifier(n_estimators=100, max_depth=4)

# Training Decision Tree classifier
clf = clf.fit(X_train, y_train)

# Predicting the response for test dataset
y_pred = clf.predict(X_test)

# Importing metrics for accuracy calculation
from sklearn import metrics

# Model Accuracy
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:" + str(accuracy))
```
